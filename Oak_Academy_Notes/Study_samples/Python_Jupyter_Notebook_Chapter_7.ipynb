```python
import pandas as pd
import numpy as np

# Örnek veri çerçevesi oluşturma
df = pd.DataFrame(np.random.rand(30, 10), columns=[f"Column_{i+1}" for i in range(10)])

# Rastgele bazı hücreleri NaN (boş) yapma
for _ in range(15):  # Toplamda 15 hücreyi boş yapalım
    row_index = np.random.randint(0, 30)  # Satır indeksi seç
    col_index = np.random.randint(0, 10)  # Sütun indeksi seç
    df.iat[row_index, col_index] = np.nan  # Seçilen hücreyi NaN (boş) yap

```

```python
df
```

```python
condition=df.isnull().any(axis=1)
```

## isnull bos deger gorme

```python
df[condition]
```

```python
# dolu verileri karşımıza getirdi
df[~condition]
```
