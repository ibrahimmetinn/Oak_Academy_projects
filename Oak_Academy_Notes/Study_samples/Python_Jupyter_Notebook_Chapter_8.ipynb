```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
note1 = pd.Series(np.random.randint(0, 100, 10), name='Note1')
note2 = pd.Series(np.random.randint(0, 100, 10), name='Note2')
note3 = pd.Series(np.random.randint(0, 100, 10), name='Note3')
name=pd.Series(data="Carolina Dennis Tom Lila Teo Melik Roam Numaxandros Yin Franklin".split())
Class=pd.Series(np.random.randint(1,8,10),name="Class")
concat_value=pd.Series(data=[".class"] * 10, name="Concat_Value")
Concatenate_class = pd.Series((Class.astype(str) + concat_value),name="Class")


School=pd.Series(np.array([np.nan,"Robert College","London Big High School",np.nan,"Deeper College",np.nan,np.nan,"Ours Lady","Destination Catholic",np.nan]),name="School")

concatenated_series = pd.concat([note1, note2, note3, Concatenate_class, School],axis=1)
df = pd.DataFrame({'Name': name, 'Note1': note1, 'Note2': note2, 'Note3': note3, 'Class': Concatenate_class, 'School': School})
```

```python
df_copy["Mean_Note"]=df.loc[:, "Note1":"Note3"].mean(axis=1)
```
